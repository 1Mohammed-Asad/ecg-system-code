```python
# =============================================================================
# CLINICAL-GRADE ECG ARRHYTHMIA CLASSIFICATION SYSTEM
#
# Description:
# This script implements a complete, end-to-end deep learning pipeline for
# classifying ECG heartbeats. It is based on the comprehensive research and
# architectural design outlined previously, aiming for clinical-grade accuracy
# (>90%) and robustness.
#
# The pipeline includes:
# 1. Data Loading: Ingests and combines the MIT-BIH and PTBDB datasets.
# 2. Preprocessing: Applies advanced filtering, resampling, and normalization.
# 3. Model Architecture: Implements a state-of-the-art hybrid 1D-CNN + LSTM model.
# 4. Training: Trains the model using class weighting to handle data imbalance.
# 5. Evaluation: Rigorously assesses performance with a full suite of metrics.
# 6. Deployment: Shows how to save the model and use it for inference.
#
# To Run:
# 1. Ensure you have the following files in the same directory as this script:
#    - mitbih_test.csv
#    - ptbdb_abnormal.csv
#    - ptbdb_normal.csv
#    - mitbih_train.csv (This is the corresponding training set for mitbih_test.csv)
# 2. Install the required libraries:
#    pip install pandas numpy tensorflow scikit-learn matplotlib
# 3. Execute the script in a Python environment (e.g., a Jupyter Notebook cell).
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Deep Learning and Machine Learning Libraries ---
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import (Input, Conv1D, BatchNormalization, ReLU, 
                                     MaxPooling1D, Bidirectional, LSTM, 
                                     GlobalMaxPooling1D, Dropout, Dense)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
from sklearn.preprocessing import LabelEncoder

# --- Signal Processing Libraries ---
from scipy.signal import resample

# --- Configuration and Setup ---
# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define class labels based on the AAMI standard for MIT-BIH
AAMI_CLASSES = {
    0: 'Normal (N)',
    1: 'Supraventricular Ectopic (S)',
    2: 'Ventricular Ectopic (V)',
    3: 'Fusion (F)',
    4: 'Unclassifiable (Q)'
}
print("âœ… System Initialized: All libraries loaded and configurations set.")


# =============================================================================
# PHASE 1: DATA CURATION & UNIFICATION
# =============================================================================
print("\n--- PHASE 1: Loading and Unifying ECG Datasets ---")

def load_and_prepare_data():
    """
    Loads, combines, and prepares the MIT-BIH and PTBDB datasets.
    This function handles the initial data ingestion and structuring.
    """
    # --- Load MIT-BIH Arrhythmia Data ---
    # We need both train and test sets for a proper workflow.
    # Assuming 'mitbih_train.csv' is available alongside 'mitbih_test.csv'.
    try:
        mit_train_df = pd.read_csv('mitbih_train.csv', header=None)
        mit_test_df = pd.read_csv('mitbih_test.csv', header=None)
        mit_df = pd.concat([mit_train_df, mit_test_df], axis=0)
        print(f"Successfully loaded MIT-BIH data. Shape: {mit_df.shape}")
    except FileNotFoundError:
        print("Error: 'mitbih_train.csv' not found. Please ensure it's in the directory.")
        # As a fallback, we'll just use the test set, but this is not ideal for training.
        try:
            mit_df = pd.read_csv('mitbih_test.csv', header=None)
            print("Warning: Using only 'mitbih_test.csv'. Model performance will be limited.")
        except FileNotFoundError:
            print("CRITICAL ERROR: 'mitbih_test.csv' also not found. Cannot proceed.")
            return None, None

    # --- Load PTB Diagnostic ECG Data (PTBDB) ---
    try:
        ptb_abnormal_df = pd.read_csv('ptbdb_abnormal.csv', header=None)
        ptb_normal_df = pd.read_csv('ptbdb_normal.csv', header=None)
        ptb_df = pd.concat([ptb_normal_df, ptb_abnormal_df], axis=0)
        print(f"Successfully loaded PTBDB data. Shape: {ptb_df.shape}")
    except FileNotFoundError:
        print("CRITICAL ERROR: PTBDB CSV files not found. Cannot proceed.")
        return None, None

    # --- Data Unification and Label Harmonization ---
    # Separate features (X) and labels (y)
    X_mit = mit_df.iloc[:, :-1].values
    y_mit = mit_df.iloc[:, -1].astype(int).values

    X_ptb = ptb_df.iloc[:, :-1].values
    y_ptb = ptb_df.iloc[:, -1].astype(int).values
    
    # Map PTBDB binary labels to the AAMI scheme
    # 0 (Normal) in PTBDB maps to 0 (Normal) in AAMI
    # 1 (Abnormal) in PTBDB will be mapped to 2 (Ventricular Ectopic), a common and critical abnormal class
    y_ptb_mapped = np.where(y_ptb == 1, 2, 0)

    # Combine the datasets
    X_combined = np.concatenate((X_mit, X_ptb), axis=0)
    y_combined = np.concatenate((y_mit, y_ptb_mapped), axis=0)
    
    print(f"âœ… Datasets successfully unified. Total samples: {len(X_combined)}")
    return X_combined, y_combined

X, y = load_and_prepare_data()

if X is None:
    print("\nHalting execution due to data loading errors.")
else:
    # =============================================================================
    # PHASE 2: SIGNAL PREPROCESSING
    # =============================================================================
    print("\n--- PHASE 2: Preprocessing Signals for Model Ingestion ---")
    
    def preprocess_signals(X_data):
        """
        Applies resampling and normalization to the dataset.
        The provided CSVs are already filtered and segmented, so we focus on
        ensuring uniform length and amplitude scaling.
        """
        # --- Temporal Harmonization: Resampling to a Uniform Length ---
        # The model requires all inputs to have the same length. We'll resample
        # each heartbeat to a fixed length of 187, standard for these datasets.
        TARGET_LENGTH = 187
        X_resampled = np.array([resample(x, TARGET_LENGTH) for x in X_data])
        print(f"Resampled all signals to a uniform length of {TARGET_LENGTH}.")

        # --- Amplitude Harmonization: Z-Score Normalization ---
        # This is done on a per-beat basis to focus on morphology.
        for i in range(X_resampled.shape[0]):
            mean = np.mean(X_resampled[i])
            std = np.std(X_resampled[i])
            if std > 0: # Avoid division by zero for flat signals
                X_resampled[i] = (X_resampled[i] - mean) / std
        print("Applied Z-score normalization to each heartbeat.")

        # --- Reshape for CNN Input ---
        # The CNN layer expects a 3D tensor: (samples, timesteps, channels)
        X_processed = np.expand_dims(X_resampled, axis=-1)
        print(f"âœ… Preprocessing complete. Final data shape: {X_processed.shape}")
        return X_processed

    X_processed = preprocess_signals(X)

    # --- Data Splitting ---
    # Stratified split to maintain class distribution in train and test sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_processed, y, test_size=0.2, random_state=42, stratify=y
    )
    print(f"Data split into training ({len(X_train)}) and testing ({len(X_test)}) sets.")


    # =============================================================================
    # PHASE 3: MODEL ARCHITECTURE
    # =============================================================================
    print("\n--- PHASE 3: Building the Hybrid CNN-LSTM Deep Learning Model ---")

    def build_ecg_classifier(input_shape, num_classes):
        """
        Builds the state-of-the-art hybrid 1D-CNN and Bidirectional LSTM model.
        """
        model = Sequential(name="Clinical_Grade_ECG_Classifier")
        
        # Input Layer
        model.add(Input(shape=input_shape, name="Input_ECG_Signal"))

        # --- Convolutional Feature Extraction Blocks ---
        # Block 1
        model.add(Conv1D(filters=64, kernel_size=5, strides=1, padding='causal', name="Conv1D_1"))
        model.add(BatchNormalization(name="BatchNorm_1"))
        model.add(ReLU(name="ReLU_1"))
        model.add(MaxPooling1D(pool_size=2, strides=2, padding='same', name="MaxPool_1"))

        # Block 2
        model.add(Conv1D(filters=128, kernel_size=5, strides=1, padding='causal', name="Conv1D_2"))
        model.add(BatchNormalization(name="BatchNorm_2"))
        model.add(ReLU(name="ReLU_2"))
        model.add(MaxPooling1D(pool_size=2, strides=2, padding='same', name="MaxPool_2"))
        
        # Block 3
        model.add(Conv1D(filters=256, kernel_size=5, strides=1, padding='causal', name="Conv1D_3"))
        model.add(BatchNormalization(name="BatchNorm_3"))
        model.add(ReLU(name="ReLU_3"))
        model.add(MaxPooling1D(pool_size=2, strides=2, padding='same', name="MaxPool_3"))

        # --- Recurrent Temporal Analysis Block ---
        model.add(Bidirectional(LSTM(64, return_sequences=False), name="Bidirectional_LSTM"))
        model.add(Dropout(0.3, name="Dropout_1")) # Regularization

        # --- Output Classification Block ---
        model.add(Dense(128, activation='relu', name="Dense_1"))
        model.add(Dropout(0.3, name="Dropout_2")) # Regularization
        model.add(Dense(num_classes, activation='softmax', name="Output_Softmax"))

        # Compile the model
        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy', # Use this for integer labels
            metrics=['accuracy']
        )
        return model

    input_shape = (X_train.shape[1], 1)
    num_classes = len(AAMI_CLASSES)
    model = build_ecg_classifier(input_shape, num_classes)
    
    print("âœ… Model architecture successfully built.")
    model.summary()


    # =============================================================================
    # PHASE 4: MODEL TRAINING
    # =============================================================================
    print("\n--- PHASE 4: Training the Model with Advanced Callbacks ---")

    # --- Address Class Imbalance with Class Weights ---
    # This is critical for learning the rare but important arrhythmia classes.
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weight_dict = dict(enumerate(class_weights))
    print("Calculated class weights to handle data imbalance:")
    for i, w in class_weight_dict.items():
        print(f"  - Class {AAMI_CLASSES[i]}: Weight {w:.2f}")

    # --- Define Callbacks for Smart Training ---
    # EarlyStopping: Prevents overfitting by stopping when performance plateaus.
    early_stopping = EarlyStopping(
        monitor='val_loss', patience=10, restore_best_weights=True, verbose=1
    )
    # ReduceLROnPlateau: Reduces learning rate for fine-tuning when loss stagnates.
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001, verbose=1
    )

    # --- Start Training ---
    print("\nStarting model training... (This may take several minutes)")
    history = model.fit(
        X_train, y_train,
        batch_size=128,
        epochs=100, # High number, but EarlyStopping will find the optimal point
        validation_split=0.2,
        class_weight=class_weight_dict,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    print("âœ… Model training complete.")


    # =============================================================================
    # PHASE 5: PERFORMANCE EVALUATION
    # =============================================================================
    print("\n--- PHASE 5: Evaluating Model Performance on Unseen Test Data ---")

    # --- Generate Predictions ---
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)

    # --- Display Comprehensive Performance Report ---
    print("\n==========================================================")
    print("           Clinical-Grade Performance Report            ")
    print("==========================================================")
    
    # Overall Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nğŸ¯ Overall Test Accuracy: {accuracy * 100:.2f}%\n")
    
    if accuracy > 0.90:
        print("ğŸ‰ SUCCESS: Model has exceeded the 90% accuracy target!")
    else:
        print("INFO: Model performance is below the 90% target. Further tuning may be needed.")

    # Detailed Classification Report (Precision, Recall, F1-Score)
    print("\nğŸ“Š Detailed Classification Metrics:\n")
    target_names = [AAMI_CLASSES[i] for i in range(num_classes)]
    print(classification_report(y_test, y_pred, target_names=target_names))

    # Confusion Matrix
    print("\nğŸ” Confusion Matrix:\n")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()


    # =============================================================================
    # PHASE 6: DEPLOYMENT PATHWAY
    # =============================================================================
    print("\n--- PHASE 6: Saving Model and Preparing for Deployment ---")

    # --- Save the Trained Model ---
    MODEL_FILENAME = 'ecg_arrhythmia_classifier_v1.h5'
    model.save(MODEL_FILENAME)
    print(f"\nâœ… Trained model saved successfully as '{MODEL_FILENAME}'")

    # --- Define Inference Function for Deployment ---
    def predict_ecg_arrhythmia(ecg_beat_signal, model_path=MODEL_FILENAME):
        """
        Takes a single pre-segmented ECG beat, preprocesses it, and returns a prediction.
        This function simulates how the model would be used in a real device.
        """
        # 1. Load the pre-trained model
        try:
            deployed_model = load_model(model_path)
        except Exception as e:
            return {"error": f"Could not load model: {e}"}

        # 2. Preprocess the input signal
        # Ensure input is a numpy array
        signal = np.array(ecg_beat_signal).flatten()
        
        # Resample to the target length the model was trained on
        processed_signal = resample(signal, 187)
        
        # Z-score normalization
        mean = np.mean(processed_signal)
        std = np.std(processed_signal)
        if std > 0:
            processed_signal = (processed_signal - mean) / std
            
        # Reshape for model prediction (1 sample, 187 timesteps, 1 channel)
        model_input = np.expand_dims(np.expand_dims(processed_signal, axis=0), axis=-1)

        # 3. Make a prediction
        prediction_probs = deployed_model.predict(model_input)[0]
        predicted_class_index = np.argmax(prediction_probs)
        confidence = prediction_probs[predicted_class_index]

        # 4. Format the output
        result = {
            "predicted_class": AAMI_CLASSES[predicted_class_index],
            "confidence": f"{confidence:.2%}",
            "full_probabilities": {AAMI_CLASSES[i]: f"{p:.2%}" for i, p in enumerate(prediction_probs)}
        }
        return result

    # --- Example of using the inference function ---
    print("\n--- Demonstrating Inference on a Single Test Sample ---")
    # Take the first sample from the test set for demonstration
    sample_beat = X_test[0].flatten() 
    prediction = predict_ecg_arrhythmia(sample_beat)
    
    print("\n--- ECG SCANNER ANALYSIS ---")
    print(f"True Label: {AAMI_CLASSES[y_test[0]]}")
    print(f"Prediction: {prediction['predicted_class']}")
    print(f"Confidence: {prediction['confidence']}")
    print("----------------------------")
    print("Full Probability Distribution:")
    for cls, prob in prediction['full_probabilities'].items():
        print(f"  - {cls}: {prob}")
    print("----------------------------")

```
